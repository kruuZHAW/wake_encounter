```python
forces = ["C_L", "C_D", "C_Y"] #Lift, Drag, Side
moments = ["C_l", "C_m", "C_n"] #rolling, pitching, yawing

# List of dataframes
dataframes = [result1, result2, result3, result4, result5, result6]
labels = ["0deg_1s", "0deg_75s", "45deg_75s", "90deg_1s", "90deg_75s", "Random parameters"]
# colormap = cm.get_cmap('viridis', len(dataframes))
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

plt.figure(figsize=(15, 8))
plt.grid(alpha=0.3, linestyle='--')

for i, (df, label) in enumerate(zip(dataframes, labels)):
    # color = colormap(i)
    plt.plot(df["C_m"], label=label, linewidth=2)

plt.legend(title="Scenarios", fontsize='medium', frameon=False)
plt.legend()

# Add labels and title for better clarity
plt.xlabel("Time (s)", fontsize=12)
plt.ylabel("Pitching Moment", fontsize=12)
plt.title("Comparison of pitching moments", fontsize=14, fontweight='bold')

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

plt.gca().spines['top'].set_visible(False) 
plt.gca().spines['right'].set_visible(False) 

# Show the plot
plt.show()
```
